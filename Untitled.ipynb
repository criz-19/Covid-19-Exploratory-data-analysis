```python

# --- Create required columns ---
df['vaccination_rate'] = (df['people_vaccinated'] / df['population']) * 100
df['death_rate'] = (df['total_deaths'] / df['total_cases']) * 100

# --- Create vaccination groups ---
df['vaccination_group'] = pd.cut(
    df['vaccination_rate'],
    bins=[0, 25, 50, 75, 100],
    labels=['0-25%', '25-50%', '50-75%', '75-100%']
)

# --- Remove invalid values ---
df = df.dropna(subset=['vaccination_group', 'death_rate'])

# --- Group data ---
grouped = df.groupby('vaccination_group', observed=False)['death_rate'].mean().reset_index()

# --- Plot Bar Graph ---
plt.figure(figsize=(10,6))

sns.barplot(
    x='vaccination_group',
    y='death_rate',
    hue='vaccination_group',
    data=grouped,
    palette='viridis',
    legend=False
)

plt.title("Average Death Rate by Vaccination Level")
plt.xlabel("Vaccination Rate Groups")
plt.ylabel("Average Death Rate (%)")

# Add values on top
for i, v in enumerate(grouped['death_rate']):
    plt.text(i, v, f"{v:.2f}", ha='center')

plt.show()
```
